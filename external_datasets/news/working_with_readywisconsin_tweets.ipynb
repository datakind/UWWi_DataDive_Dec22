# import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# discover @ReadyWisconsin tweets

```python
tweets = pd.read_csv("/content/ReadyWiscEmergencyAlertTweetArchive.csv", index_col=0)
tweets.info()
```

```python
tweets = tweets.drop(["_type", "user", "media", "coordinates", "place", "cashtags", "retweetedTweet"], axis=1)
```

```python
tweets.isna().sum()
```

```python
tweets["date"] = pd.to_datetime(tweets["date"])
```

## 10.000 tweets from 2014 to 2022, seasonality: more tweets during summer

```python
plt.figure(figsize=(16,6))
tweets["date"].hist(bins=12*9)
plt.show()
```

# loading uwwi datasets

```python
uwwi22 = pd.read_csv("https://github.com/datakind/UWWi_DataDive_Dec22/raw/main/uwwi_datasets/uwwi_dataset_interactions/uwwi_dataset_interactions_2022.csv")
uwwi21 = pd.read_csv("https://github.com/datakind/UWWi_DataDive_Dec22/raw/main/uwwi_datasets/uwwi_dataset_interactions/uwwi_dataset_interactions_2021.csv")
uwwi20 = pd.read_csv("https://github.com/datakind/UWWi_DataDive_Dec22/raw/main/uwwi_datasets/uwwi_dataset_interactions/uwwi_dataset_interactions_2020.csv")

uwwi = pd.DataFrame()
uwwi = pd.concat([uwwi20, uwwi21, uwwi22])

uwwi["Interaction_CreateStamp"] = pd.to_datetime(uwwi.Interaction_CreateStamp, utc=True)
uwwi["Interaction_EditStamp"] = pd.to_datetime(uwwi.Interaction_EditStamp, utc=True)

uwwi.head()
```

# @ReadyWisconsin & UWWi Calls

```python
plt.figure(figsize=(14,10))
uwwi.Interaction_CreateStamp.hist(bins=365*3)
tweets["date"][tweets['date'] > "2020-01-01"].hist(bins=3*365)
plt.legend(labels=["Calls UWWi", "Tweets @ReadyWisconsin"])
plt.show()
```

## on average 3 tweets/day, max 57/d

```python
tweets20 = tweets[tweets['date'] > "2020-01-01"]
tweets20_count = tweets20["url"].groupby(tweets20['date'].dt.floor('d')).count()
tweets20_count.describe()
```

```python
tweets20_count
```

```python
calls = uwwi.Interaction_CreateStamp.groupby(uwwi["Interaction_CreateStamp"].dt.floor('d')).count()
```

```python
df1 = pd.DataFrame(calls)
df2 = pd.DataFrame(tweets20_count)
calls_tweets = df1.join(df2)
calls_tweets.rename({'Interaction_CreateStamp': 'calls', 'url': 'tweets'}, axis=1, inplace=True)
calls_tweets.index.names = ['day']
calls_tweets.tweets.fillna(0, inplace=True)
calls_tweets
```

## low correlation between calls & tweets

```python
calls_tweets.corr()
```

## calls on days with more tweets

```python
for i in range(10, 60, 10):
    cutoff = i

    anl_days = calls_tweets[[day in tweets20_count[tweets20_count>cutoff].index for day in calls_tweets.index]]
    # print("days with >", cutoff, " tweets:", len(anl_days), "; mean of calls on these days:", anl_days.calls.mean())

    nml_days = calls_tweets[[day not in tweets20_count[tweets20_count>cutoff].index for day in calls_tweets.index]]
    # print("days with <", cutoff, " tweets:", len(nml_days), "; mean of calls on these days:", nml_days.calls.mean())

    x = anl_days["tweets"]
    y = anl_days["calls"]
    plt.scatter(x, y)
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m*x+b)
    print(m, b)
    plt.show()

    print(cutoff, anl_days.corr().iloc[1,0])
```

```python
cutoff = 30

anl_days = calls_tweets[[day in tweets20_count[tweets20_count>cutoff].index for day in calls_tweets.index]]
anl_days["anl"] = [100 for _ in range(len(anl_days))]
anl_days.drop(["calls", "tweets"], axis=1, inplace=True)
df = pd.merge(calls_tweets, anl_days, how='outer', left_index=True, right_index=True)
df["anl"].fillna(0, inplace=True)

plt.figure(figsize=(16,8))
plt.plot(df)
plt.show()
```

```python
anl_tweets = tweets[["date", "content"]][[day in anl_days.index for day in tweets.date.dt.floor("d")]]
anl_tweets.date.dt.floor("d").nunique()
```

# create interactive timeline

```python
calls_tweets = calls_tweets.reset_index(level=0)
calls_tweets
```

```python
import plotly.express as px
```

```python
snwd = pd.read_csv("https://github.com/bsenst/UWWi_DataDive_Dec22/raw/main/external_datasets/news/calls_per_day.csv")
snwd
```

```python
df = pd.merge(calls_tweets, snwd["snwd"], how='outer', left_index=True, right_index=True)

ess_tweets = []
for row in anl_tweets.iterrows():
  ess_tweets.append(str(row[1][0])[:10] + " " + " ".join(str(row[1][1]).split()[:4]))
ess_tweets = set(ess_tweets)

ess_tweets = {'2020-07-19 ReadyWisconsin Alert Severe Thunderstorm',
 '2020-08-25 ReadyWisconsin Alert Severe Thunderstorm',
 '2021-07-29 ReadyWisconsin Alert Severe Thunderstorm',
 '2021-08-11 Extreme heat being felt',
 '2022-06-15 Potentially dangerous storms will',
 '2022-08-03 ReadyWisconsin Alert Flash Flood',
}
```

```python
fig = px.line(df, x='day', y=["calls", "tweets", "snwd"])

for tw in ess_tweets:
  fig.add_annotation(x=tw[:10], y=80,
              text=tw[11:],
              showarrow=True,
              textangle=-45,
              arrowhead=1)

# fig.write_html("calls_tweets_snwd.html")
fig.show()
```
